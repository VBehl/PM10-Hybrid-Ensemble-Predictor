```python
# Step 1: Data Understanding and Preprocessing
# --------------------------------------------

# 📦 Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

# 🧪 Settings for plots
plt.style.use('seaborn-v0_8')  # Updated for compatibility
plt.rcParams['figure.figsize'] = (14, 6)

# 📂 Load dataset
file_path = "./Dataset/Pune_SmartCity_Test_Dataset.csv"
df = pd.read_csv(file_path)

# 🔍 Preview data
print("Dataset Shape:", df.shape)
display(df.head())
display(df.info())
display(df.describe())

# 📅 Parse timestamp and set as index
df['timestamp'] = pd.to_datetime(df['LASTUPDATEDATETIME'])
df.drop(columns='LASTUPDATEDATETIME', inplace=True)
df = df.sort_values('timestamp')
df.set_index('timestamp', inplace=True)

# 🧹 Handle missing values
print("\nMissing Values (before cleaning):")
print(df.isnull().sum())

# Fill using forward fill and backward fill
df = df.fillna(method='ffill').fillna(method='bfill')
df = df.dropna()

print("\nMissing Values (after cleaning):")
print(df.isnull().sum())

# 🧾 Feature selection (drop non-numeric and location-related info if needed)
non_features = ['NAME', 'Lattitude', 'Longitude']  # You can modify this list if needed
df_model = df.drop(columns=non_features)

# 📊 Correlation matrix to understand feature relationships
plt.figure(figsize=(16, 10))
sns.heatmap(df_model.corr(), annot=False, cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()

# 📏 Normalize the data using MinMaxScaler
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(
    scaler.fit_transform(df_model),
    columns=df_model.columns,
    index=df_model.index
)

import os

# ✅ Ensure the 'Preprocessed' directory exists
output_dir = "./Preprocessed"
os.makedirs(output_dir, exist_ok=True)

# 💾 Save preprocessed data to CSV
output_file = os.path.join(output_dir, "preprocessed_air_quality_data.csv")
df_scaled.to_csv(output_file)
print(f"✅ Preprocessed data saved as '{output_file}'")


# ✂️ Train-test split (time-based)
train_size = int(len(df_scaled) * 0.8)
train_data = df_scaled.iloc[:train_size]
test_data = df_scaled.iloc[train_size:]

print(f"\nTrain data shape: {train_data.shape}")
print(f"Test data shape: {test_data.shape}")

# 📈 Visualize PM2.5 over time (Train vs Test)
pm25_cols = [col for col in df_scaled.columns if 'PM2' in col.upper()]
if pm25_cols:
    pm2_col = pm25_cols[0]  # Use the first matching PM2 column
    plt.figure(figsize=(14, 5))
    plt.plot(train_data.index, train_data[pm2_col], label=f"{pm2_col} - Train")
    plt.plot(test_data.index, test_data[pm2_col], label=f"{pm2_col} - Test", alpha=0.7)
    plt.title(f"{pm2_col} Trend Over Time")
    plt.xlabel("Timestamp")
    plt.ylabel(f"Normalized {pm2_col}")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("⚠️ No PM2.x column found to plot.")
```
